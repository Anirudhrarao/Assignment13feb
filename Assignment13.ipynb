### Solution (3) 
* Arithmetic error are simply raised when we do numeric calculations.
#### There are 3 types of errors in Arithmetic error:
1. OverFlowError
2. ZeroDivisionError
3. FloatingPointError

```python
try:
    120/0
except ArithmeticError as e:
    print(e)
```

```python
import math
try:
    print(math.exp(1000))
except OverflowError as e:
    print(e)
```

### Solution (4)
* A lookuperror is used when we need to handle exceptions like indexerror and keyerror.

```python
x = [1,2,3,4]
try:
    print(x[10])
except LookupError as e:
    print(e,e.__class__)
```
